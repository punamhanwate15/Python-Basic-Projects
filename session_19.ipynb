With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
print("Minimum House Value:", df["median_house_value"].min())
print("Maximum House Value:", df["median_house_value"].max())
print("Mean House Value:", df["median_house_value"].mean())
print("Median House Value:", df["median_house_value"].median())
```

```
# 1. The median_house_value ranges from 14999 to 500001.
# 2. The mean house value is higher than the median, indicating a slightly right-skewed distribution.
# 3. The dataset contains a wide range of house prices, showing high variability.
# 4. Most numeric features have different scales, so feature scaling may be required before model training
```

```
cols = ["median_income", "housing_median_age",
        "total_rooms", "median_house_value"]

plt.figure(figsize=(8,6))

for i, col in enumerate(cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Histogram of {col}")

plt.tight_layout()
plt.show()
```

```
# 1. Median Income:
# Most values are between 2 and 6, showing that most areas have moderate income levels.

# 2. Housing Median Age:
# Most houses are 15 to 35 years old, while very old houses are less common.

# 3. Total Rooms:
# The distribution is right-skewed, with a few houses having a very large number of rooms.

# 4. Median House Value:
# Most house prices are in the middle range, and the maximum value is capped at 500001.
```

```
# Create categorical columns
df["Income_Level"] = pd.cut(df["median_income"],
                            bins=[0,2,4,6,15],
                            labels=["Low","Medium","High","Very High"])

df["House_Age"] = pd.cut(df["housing_median_age"],
                         bins=[0,20,40,60],
                         labels=["Young","Middle","Old"])

plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
sns.countplot(data=df, x="ocean_proximity")
plt.xticks(rotation=30)

plt.subplot(2,2,2)
sns.countplot(data=df, x="Income_Level")

plt.subplot(2,2,3)
sns.countplot(data=df, x="House_Age")

plt.tight_layout()
plt.show()
```

```
# 1. The "<1H OCEAN" category has the highest number of houses,
#    showing that many residential areas are located near the ocean.

# 2. Most houses fall under the Medium income category,
#    indicating that the majority of households have moderate income levels.

# 3. The Middle age (20–40 years) category has the highest count,
#    suggesting that most houses in the dataset are neither very new nor very old.
```

```
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```
# 1. median_income has the strongest positive correlation with median_house_value.
# 2. housing_median_age shows a moderate positive correlation with house value.
# 3. longitude has a negative correlation with the target variable.
# 4. Most other features have weak to moderate correlations with house prices.
```

```
# Before Encoding
print("Before Encoding:")
print(df[["ocean_proximity"]].head())

# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=["ocean_proximity"], dtype=int)

# After Encoding
print("\nAfter Encoding:")
print(df_encoded.filter(like="ocean_proximity").head())
```

```
# 1. The 'ocean_proximity' column is categorical and cannot be used directly.
# 2. One-Hot Encoding converts each category into separate binary (0/1) columns.
# 3. This allows machine learning models to process categorical data correctly.
# 4. The original categorical column is replaced with new encoded columns.
```
